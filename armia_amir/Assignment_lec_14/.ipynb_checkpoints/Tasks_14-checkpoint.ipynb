```python
df_1= pd.read_csv("ufo.csv")
df_1
```

```python
df_1['Time']=pd.to_datetime(df_1['Time'], format='%m/%d/%Y %H:%M')
```

```python
df_1
```

```python
def month(row):
    if row["Time"].month == 12 :
        return 1
    else:
        return 0
```

```python
df_1["IN_December"] = df_1.apply(month,axis=1)
df_1["Time_Diff_From_WW1"] = abs(datetime(1914,7,28) - df_1["Time"]) / np.timedelta64(1,"Y")
df_1
```

```python
df = pd.read_csv('Restaurant_Reviews.csv', sep='\t')
df
```

```python
reviews = df['Review']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
reviews_vecctor = vectorizer.fit_transform(reviews)
reviews_vecctor.toarray()
```
