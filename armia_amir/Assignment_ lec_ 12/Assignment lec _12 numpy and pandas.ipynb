```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('bikes.csv')
df
```

```python
df.info()
```

```python
### Make Profit Feature.

def calculate_profit(x):
    casual_customers =x['casual']
    registerd_customers =x['registered']
    registerd_price_per_hour = 5
    casual_price_per_hour = 20
    taxes= 0.14 
    maintance_per_year=1500/(12*30*24)
    profit_=taxes*(casual_customers*(casual_price_per_hour+maintance_per_year)+registerd_customers*(casual_price_per_hour+maintance_per_year))
    return profit_
df['profit'] = df[['casual', 'registered']].apply(calculate_profit, axis=1)  
df
```

```python
### Distribution rental_bikes_count & profit.
df['profit'].describe()
```

```python
plt.hist(df['profit'])
plt.show()
```

```python
df['rented_bikes_count'].describe()
```

```python
plt.hist(df['rented_bikes_count'])
plt.show()
```

```python
### Profit for each day.
df['day']=df['datetime'].apply( lambda x : x.split()[0])
df.groupby('day').sum()['profit']
```

```python
### Which season that has more registeration & profit.
df.groupby('season').sum()['profit']
```

```python
df.groupby('season').sum()['registered']
```

```python
### Which weather condition that has more registeration & profit.
df.groupby('weather').sum()['registered']
```

```python
df.groupby('weather').sum()['profit']
```

```python
### Correlation between profit & bikes count with all features.

df[['profit','rented_bikes_count']].corr()
```

```python
### Rentals during rush hours (7-9 am & 3-5 pm).
df['Rentals_HOURS']=df['datetime'].apply(lambda x : int(x.split()[1][0:2]))
df[(df['Rentals_HOURS'].between(7,9,inclusive=True)) | (df['Rentals_HOURS'].between(15,17,inclusive=True))].groupby('Rentals_HOURS').sum()['rented_bikes_count']
```

```python
###  Registered or casual is more.
df[['casual','registered']].sum()
```

```python
### Rentals during workday & holiday.
df['rented_bikes_count'][(df['workingday']==1) | (df['holiday']==1)]
```

```python
df['formatted_date']= pd.to_datetime(df['datetime'])
df['week_of_year'] = df.formatted_date.apply(lambda x: x.weekofyear)
df.groupby('week_of_year').mean()['rented_bikes_count'].mean()
```

```python
def _day (x):
    if '-12-09' in x :
        return True
    elif '-06-02' in x:
        return True
    else : 
        return False
df[df['day'].apply(_day)].groupby('day').sum()['profit']
```
