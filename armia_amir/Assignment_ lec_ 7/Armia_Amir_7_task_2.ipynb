```python
import requests
respons=requests.get('https://www.metaweather.com/api/location/1521894/')
respons.content
```

```python
data=respons.json()
data
```

```python
 data['consolidated_weather'][0]['the_temp'], data['consolidated_weather'][0]['humidity'], data['consolidated_weather'][0]['visibility'], data['consolidated_weather'][0]['air_pressure'], data['consolidated_weather'][0]['wind_speed'], data['consolidated_weather'][0]['created']
```

```python
for c in data['consolidated_weather']:
    print(f"temperature: {c['the_temp']}, humidity: {c['humidity']}, visibility: {c['visibility']}, air pressure :{c['air_pressure']}, wind speed: {c['wind_speed']}, datetime: {c['applicable_date']}")
```

```python
import csv
```

```python
with open('day_data.csv','w',newline='') as f:
    writer=csv.DictWriter(f,fieldnames=['Temperature','Humidity','Visibility','Air_Pressure','Wind_Speed','Date'])
    writer.writeheader()
    for dat in data['consolidated_weather']:
        temperature=dat['the_temp']
        humidity=dat['humidity']
        visibility=dat['visibility']
        air_Pressure=dat['air_pressure']
        wind_Speed=dat['wind_speed']
        created=dat['applicable_date']
        writer.writerow({'Temperature':temperature,'Humidity':humidity,'Visibility':visibility,'Air_Pressure':air_Pressure,'Wind_Speed':wind_Speed,'Date':created})
```
