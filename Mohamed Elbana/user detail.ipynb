```python
for users in dic['data']:
     print(f"id:{users['id']},name: {users['name']} , email: {users['email']} , gender: {users['gender']}, status :{users['status']}  ")
```

```python
import csv
```

```python
with open('user.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['id','name','email','gender','status'])
    for user in dic['data']:
        id=user['id']
        name=user['name']
        email=user['email']
        gender=user['gender']
        status=user['status']   
        writer.writerow([id,name,email,gender,status])
       
          
```
