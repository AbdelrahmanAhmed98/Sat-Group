```python
import csv
import pandas as pd
```

## Day1

```python
with open('weather.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['day 1'])
    writer.writerow(['id','temperature', 'humidity', 'visibility', 'air pressure', 'wind speed','datetime'])
    for i in data['consolidated_weather']:
        id_num=i['id']
        temp=i['the_temp']
        humidity=i['humidity']
        visibility=i['visibility']
        air_pressure=i['air_pressure']
        wind_speed=i['wind_speed']
        created=i['created']
        writer.writerow([id_num,temp,humidity,visibility,air_pressure,wind_speed,created])
dataframe=pd.read_csv('weather.csv')
dataframe
```

```python
with open('user_data.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['name', 'email', 'gender', 'Status'])
    for user in data['data']:
        name=user['name']
        email=user['email']
        gender=user['gender']
        status=1 if user['status']=='active' else 0
        writer.writerow([name, email, gender, status])
```
