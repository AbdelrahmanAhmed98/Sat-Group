## 1) Using ufo.csv dataset, extract Month from &#39;Time&#39; and use it to create new column and call it &#39;IN_December&#39;, then it will have a value of &#39;1&#39; if the month is December and &#39;0&#39; otherwise.

```python
df=pd.read_csv('ufo.csv')
df.head()
```

```python
df['Time']=pd.to_datetime(df['Time'],format='%m/%d/%Y %H:%M')
```

```python
def month_checker(x):
    if x['Time'].month ==12:
        return 1
    else:
        return 0
df['IN_December']=df.apply(month_checker ,axis=1)
df
```

```python
df['time_diff_from_WW1']=abs(datetime(1914,7,28)-df['Time'])/np.timedelta64(1,'Y')
df
```

## 1) Using titanic_full.csv dataset, extract Mr,Miss,Mrs,etc.. from &#39;Name&#39; column and use it to create new column and call it &#39;Title&#39;, then plot a countplot

```python
df=pd.read_csv('titanic_full.csv')
df
```

```python
def extract_title(x):
    if "Mr." in x:
        return 'Mr'
    elif 'Miss.' in x:
        return 'Miss'
    elif 'Mrs' in x:
        return 'Mrs'
    else:
        return 'untitled'
df['title']=df['Name'].apply(extract_title)
df
```

```python
sns.countplot('title',data=df)
```

## Using visitor-interests.csv dataset, from the first 50 row extract Browser, OS, Device data from &#39;UserAgent&#39; column and use it to create new columns for Browser, OS, Device.

```python
df=pd.read_csv('visitor-interests.csv')
df=df.head(50)
df
```

```python
df['UserAgent']
```

```python
import user_agents
```

```python
def extractor(x):
    ua=user_agents.parse(x)
    return pd.Series((ua.browser.family,ua.os.family,ua.device.family))
df[['Browser','OS','Device']]=df['UserAgent'].apply(extractor)
df
```

## 2) Using visitor-interests.csv dataset, from the first 50 row extract latitude, longitude from &#39;IP&#39; column.

```python
import urllib.request
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
```

```python
def details(x):
    try:
        response = ip2geo.get(x, api_key='free')
        return pd.Series((response.latitude , response.longitude))
    except:
        return np.nan
df[['latitude','longitude']]=df['IP'].apply(details)
df
```

## Using visitor-interests.csv dataset, from the first 50 row extract latitude, longitude from &#39;IP&#39; column and use it to create new &#39;Dist_From_Egypt&#39; column calculating distance with Egypt Lat/Lonng, you can get this location from google.

```python
from geopy.distance import great_circle
```

```python
def calculate_distance(x):
    try:
        egypt=(30.033333,  31.233334)
        response = ip2geo.get(x, api_key='free')
        country = (response.latitude , response.longitude)
        return great_circle(country,egypt).kilometers
    except:
        return np.nan
df['Dist_From_Egypt']=df['IP'].apply(calculate_distance)
df
```
