```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
df=pd.read_csv('bikes.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.hist(df['rented_bikes_count'])
```

```python
sns.distplot(df['rented_bikes_count'],color='b',bins=20)
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(x='rented_bikes_count',y='registered',data=df,kind='scatter',hue='weather')
```

```python
sns.boxplot(x='season',y='rented_bikes_count',data=df)
```

```python
sns.boxplot(x='season',y='rented_bikes_count',data=df,hue='weather')
```

```python
sns.violinplot(x='season',y='rented_bikes_count',data=df)
```

```python
sns.set(rc={'figure.figsize': [7, 7]}, font_scale=1.3)
sns.stripplot(x='weather',y='rented_bikes_count',data=df,palette='viridis')
```

```python
sns.swarmplot(x='weather',y='rented_bikes_count',data=df,palette='viridis')
```

```python
sns.barplot(x='season',y='rented_bikes_count',data=df)
```

```python
sns.barplot(x='season',y='rented_bikes_count',data=df,estimator=np.median)
```

```python
sns.barplot(x='season',y='rented_bikes_count',data=df,estimator=np.max)
```

```python
sns.countplot(x='weather',data=df)
```

```python
sns.catplot(x='season',y='rented_bikes_count',data=df,kind='bar')
```

```python
sns.heatmap(df.corr())
```
