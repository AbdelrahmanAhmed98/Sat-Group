```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("50_Startups.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df,columns=['State'],drop_first=True)
df
```

```python
x=df.drop("Profit",axis=1)
y=df["Profit"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
models={
    'LR':LinearRegression(),
    'KNN':KNeighborsRegressor(n_neighbors=5),
    'SVR':SVR(),
    'DT':DecisionTreeRegressor(),
    'RF':RandomForestRegressor(n_estimators=25),
    'XGBR':XGBRegressor()
}
```

```python
for name,model in models.items():
    print(f'using:{name}')
    model.fit(x_train,y_train)
    print(f'training accuracy: {model.score(x_train,y_train)}')
    print(f'testing accuracy: {model.score(x_test,y_test)}')
    print('_'*40)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
x.columns
```

```python
d=[10000,12000,15000]+[0,0]
d=scaler.transform([d])
d
```

```python
bred=model.predict(d)
bred
```

```python
import joblib
```

```python
joblib.dump(model,'model.h5')
joblib.dump(scaler,'scaler.h5')
joblib.dump(x.columns,'features.h5')
```

```python
df=pd.read_csv('Houses prices.csv')
df
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
ct=df['city'].value_counts().head(10).index
ct
```

```python
df['city']=df['city'].apply(lambda x: x if x in ct else 'other')
df['city'].value_counts()
```

```python
df['street']=df.apply(lambda x: x['street'] if x['city'] in ct else 'other',axis=1)
df['street'].value_counts()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(['country','date'], axis=1, inplace=True)
df
```

```python
df.info()
```

```python
df=pd.get_dummies(df,columns=['statezip','street','city'],drop_first=True)
df
```

```python
x=df.drop("price",axis=1)
y=df["price"]
```

```python
x
```

```python
y
```
