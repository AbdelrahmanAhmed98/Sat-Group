```python
with open('task3.csv', 'w') as f:
    writer=csv.writer(f)
    writer.writerow(['Name', 'Email', 'Gender', 'Status'])
    for id in data['data']: 
        Name=id['name']
        Email=id['email']
        Gender=id['gender']
        if id['status']=='active':
            status=1
        elif id['status']=='inactive':
            status=0
        writer.writerow([Name, Email, Gender, status])
```
