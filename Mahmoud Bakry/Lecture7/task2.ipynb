```python
import requests
```

```python

response=requests.get('https://www.metaweather.com/api/location/1521894/')
response.content
```

```python
data=response.json() 
data
```

```python
for id in data['consolidated_weather']:
        print(f"Temperature: {id['the_temp']}, Humidity: {id['humidity']}, Visibility: {id['visibility']}")
```

```python
import csv
```

```python
with open('task2.csv', 'a') as f:
    writer=csv.writer(f)
    writer.writerow(['Temperature', 'Humidity', 'Visibility'])
    for id in data['consolidated_weather']: 
        Temperature=id['the_temp']
        Humidity=id['humidity']
        Visibility=id['visibility']
       
        writer.writerow([Temperature,Humidity,Visibility])
 
```
