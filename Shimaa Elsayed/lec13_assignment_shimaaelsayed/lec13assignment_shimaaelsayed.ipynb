## task 1 bikes.csv

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

bikes_df=pd.read_csv('bikes.csv')
bikes_df

```

### Numerical Visualization

```python
plt.plot(bikes_df['temp'],bikes_df['rented_bikes_count'],label='viz1')
plt.legend()
plt.xlabel('Temperature ..')
plt.ylabel('Rented bikes count ..')
plt.title('temp vs rented bikes count ')
```

```python
plt.plot(bikes_df['humidity'],bikes_df['rented_bikes_count'],label='viz1')
plt.legend()
plt.xlabel('humidity ..')
plt.ylabel('Rented bikes count ..')
plt.title('humidity vs rented bikes count ')
```

```python
plt.hist(bikes_df['rented_bikes_count'])
```

```python
sns.distplot(bikes_df['casual'],color='m',bins=30)#cmykrgb
```

```python
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
sns.distplot(bikes_df['registered'],kde=False,bins=30,color='g')
```

```python
sns.kdeplot(bikes_df['temp'],color='m',shade=True)
```

```python
sns.jointplot(x='temp',y='rented_bikes_count',data=bikes_df,kind='kde')
```

```python
sns.jointplot(x='humidity',y='rented_bikes_count',data=bikes_df,kind='hex')
```

```python
sns.pairplot(bikes_df)
```

```python
sns.pairplot(bikes_df,vars=['casual','registered'],height=4,aspect=1.4,hue='temp')
```

### Categorical Visualitsation

```python
sns.boxplot(y='weather',x='rented_bikes_count',data=bikes_df)
```

```python
sns.boxplot(y='season',x='rented_bikes_count',data=bikes_df)
```

```python
sns.violinplot(x='weather',y='casual',data=bikes_df)
```

```python
sns.violinplot(x='season',y='casual',data=bikes_df)
```

```python
sns.stripplot(x='weather',y='registered',data=bikes_df)
```

```python
sns.stripplot(x='weather',y='registered',data=bikes_df,hue='season')
```

```python
sns.swarmplot(x='weather',y='rented_bikes_count',data=bikes_df,hue='season',dodge=True)
```

```python
sns.barplot(x='weather',y='rented_bikes_count',data=bikes_df)
```

```python
sns.barplot(x='season',y='rented_bikes_count',data=bikes_df)
```

```python
sns.heatmap(bikes_df.corr())
```

```python
sns.countplot(x='weather',data=bikes_df)
```

```python
sns.countplot(x='season',data=bikes_df)
```

```python
sns.countplot(x='weather',data=bikes_df,hue='season')
```
