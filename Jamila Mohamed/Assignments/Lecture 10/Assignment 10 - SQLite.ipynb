```python
connection=sqlite3.connect('bookssql.db')
for i in categories:
        response=requests.get(i)
        soup=BeautifulSoup(response.text,'html.parser')
        category=soup.find('div',attrs={'class':'page-header'}).find('h1').get_text()
        for books in soup.find_all('article',attrs={'class','product_pod'}):
            name=books.find('h3').find('a').get_text()
            price=float(books.find('div',attrs={'class','product_price'}).get_text().split()[0][2:])
            rate=books.find('p').get('class')[1]
            if rate=='One':
                rate=1
            elif rate=='Two':
                rate=2
            elif rate=='Three':
                rate=3
            elif rate=='Four':
                rate=4
            elif rate=='Five':
                rate=5
            cursor=connection.execute(" INSERT INTO Books(BookName,BookPrice,BookRate,BookCategory) values(?,?,?,?) ",[name,price,rate,category])
            connection.commit()
```

```python
connection=sqlite3.connect('bookssql.db')
for i in categories:
    response=requests.get(i)
    soup=BeautifulSoup(response.text,'html.parser')
    categ=soup.find('div',attrs={'class':'page-header'}).find('h1').get_text()
    cursor=connection.execute(" INSERT INTO Categories(BookCategory) values(?) ",[categ])
    connection.commit()
```
