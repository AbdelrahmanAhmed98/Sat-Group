```python
from bs4 import BeautifulSoup
import requests
import csv
with open('bookdetails.csv','w',newline='',encoding="utf-8") as f:
    writer=csv.writer(f)
    writer.writerow(['BookName','Price','Rate','Category'])
    for i in categories:
        response=requests.get(i)
        response.text
        soup=BeautifulSoup(response.text,'html.parser')
        category=soup.find('div',attrs={'class':'page-header'}).find('h1').get_text()
        for book in soup.find_all('article',attrs={'class':'product_pod'}):
            name=book.find('h3').find('a').get('title')
            price=float(book.find('div',attrs={'class':'product_price'}).find('p',attrs={'class':'price_color'}).get_text()[2:])
            rate=book.find('p').get('class')[1]
            writer.writerow([name,price,rate,category])
```
