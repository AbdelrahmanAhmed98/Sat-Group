```python
df=pd.get_dummies(df,columns=['Age'],drop_first=False)
df
```

```python
City_Category_dict={'A':1,'B':2,'C':3,}
df['City_Category']=df['City_Category'].map(City_Category_dict)
df
```

```python
Stay_In_Current_City_Years_dict={'0':'zero','1':'one','2':'two','3':'three','4+':'four & more'}
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].map(Stay_In_Current_City_Years_dict)
df
```
