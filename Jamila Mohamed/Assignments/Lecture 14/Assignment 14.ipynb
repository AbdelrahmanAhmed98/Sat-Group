```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set(rc={'figure.figsize':[10,10]},font_scale=1.3)
```

```python
df=pd.read_csv('datasets/visitor-interests.csv')
df
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
```

```python
df=df.head(50)
df
```

```python
from geopy.distance import great_circle
```

```python
def ext_dist(x):
    try:
        ua=ip2geo.get(x,api_key='free')
        return great_circle((ua.latitude,ua.longitude),(30.044420,31.235712)).kilometers
    except:
        return 'Missing'

df['Dist_From_Egypt']=df['IP'].apply(ext_dist)
df
```

```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set(rc={'figure.figsize':[10,10]},font_scale=1.3)
```

```python
df=pd.read_csv('datasets/Restaurant_Reviews.csv',sep='\t')
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer=CountVectorizer(stop_words="english")

vector=vectorizer.fit_transform(df['Review'])

vector.toarray()
```

```python
df_resp1=pd.DataFrame(vector.toarray(),columns=vectorizer.get_feature_names())
df_resp1
```

```python
# sees the frequency of this word in this document, term frequency -> TF Term Frequency
# sees the frequency of this document in this word, term frequency -> IDF Inverse Document Frequency
# extract the necessary words -> fit
# move review by review and apply this equation on them -> transform
```
