### bikes.csv

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Assignment datasets/bikes.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#Make Profit Feature
```

```python
def ext_profit(x):
    profit=(x['registered']*5)+(x['casual']*20)
    taxes=profit*0.14
    maint=1500/(365*24)
    profit=profit-(taxes+maint)
    return profit

df['Profit']=df.apply(ext_profit,axis=1)
df
```

```python
#Distribution rental_bikes_count & profit
```

```python
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
sns.jointplot(x="rented_bikes_count",y="Profit",data=df)
```

```python
#Profit for each day
```

```python
df['Day']=df['datetime'].apply( lambda x : x.split()[0])
df.groupby('Day').sum()['Profit']
```

```python
df.groupby('season').sum()['registered']
```

```python
#Which season that has more profit?
```

```python
df.groupby('season').sum()['Profit']
```

```python
df.groupby('weather').sum()['registered']
```

```python
#Which weather condition that has more profit?
```

```python
df.groupby('weather').sum()['Profit']
```

```python
#Correlation between profit & bikes count with all features
```

```python
df.corr()[['Profit','rented_bikes_count']]
```
