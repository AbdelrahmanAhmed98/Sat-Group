### USERS DATA SCRAPE

```python
import requests
response=requests.request('GET','https://gorest.co.in/public-api/users?_format=json&access-token=H4lYmIGLXZkDlpy_w0ZVS5rgcpDXbN7fhRRy')
users_data=response.json()
users_data['data']
```

```python
import csv
with open('users.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['ID','Name','Email','Gender','Status'])
    for details in users_data['data']:
        id=details['id']
        name=details['name']
        email=details['email']
        gender=details['gender']
        if details['status']=='active':
            status=1
        elif details['status']=='inactive':
            status=0
        writer.writerow([id,name,email,gender,status])
```
