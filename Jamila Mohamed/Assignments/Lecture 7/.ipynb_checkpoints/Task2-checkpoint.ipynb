### WEATHER SCRAPE

```python
import requests
response=requests.get('https://www.metaweather.com/api/location/1521894/')
weather_data=response.json()
weather_data['consolidated_weather']
```

```python
import csv
with open('weather.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['Temperature','Humidity','Visibility','Air Pressure','Wind Speed'])
    for i in weather_data['consolidated_weather']:
        for details in i:
            temperature=i['the_temp']
            humidity=i['humidity']
            visibility=i['visibility']
            airpressure=i['air_pressure']
            windspeed=i['wind_speed']
        writer.writerow([temperature,humidity,visibility,airpressure,windspeed])
```
