```python
df=pd.get_dummies(df,columns=["AgeCategory","GenHealth"],drop_first=True,dtype=int)
```

```python
df = df.drop('Race', axis = 1)
```

```python
df.head()
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler 
num_cols = ['MentalHealth', 'BMI', 'PhysicalHealth', 'SleepTime']
Scaler = StandardScaler()
df[num_cols] = Scaler.fit_transform(df[num_cols])
```

```python
df.info()
```

```python
x = df.drop(columns =['HeartDisease'], axis = 1)

y = df['HeartDisease']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,  test_size = 0.3, random_state = 32)


```

```python
x_test.columns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
```

```python
models={
    'lr':LogisticRegression(max_iter=1000),
    #'KNn':KNeighborsClassifier(n_neighbors=5),
    'DT':DecisionTreeClassifier(),
    #'SVC':SVC(),
    'RF':RandomForestClassifier(),
    'XGBC':XGBClassifier(),
    'NB':GaussianNB()
    
}
```

```python
for name,model in models.items():
    print(f'using{name}')
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(f'Training Accuracy:{accuracy_score(y_train,model.predict(x_train))}')
    print(f'testing Accuracy:{accuracy_score(y_test,y_pred)}')
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train.values,y_train.values)
y_pred=knn.predict(x_test.values)
print(f'Training Accuracy:{accuracy_score(y_train,knn.predict(x_train.values))}')
print(f'testing Accuracy:{accuracy_score(y_test,y_pred)}')
```

```python
SVC=SVC()
SVC.fit(x_train,y_train)
y_pred=SVC.predict(x_test)
print(f'Training Accuracy:{accuracy_score(y_train,SVC.predict(x_train))}')
print(f'testing Accuracy:{accuracy_score(y_test,y_pred)}')
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(f'Training Accuracy:{accuracy_score(y_train,model.predict(x_train))}')
print(f'testing Accuracy:{accuracy_score(y_test,y_pred)}')
```
